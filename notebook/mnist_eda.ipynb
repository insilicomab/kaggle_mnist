# データの確認

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train_labels = train['label']
```

# 二次元に変換後、画像の表示

```python
train_x = train.iloc[:, 1:].values.astype('float32')
train_images = train_x.reshape(-1, 28, 28) # (Number, Height, Width)
```

```python
# 画像の表示
fig, ax = plt.subplots(1, 3, figsize=(20, 10))
ax[0].imshow(train_images[0], cmap='gray')
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[1].imshow(train_images[1], cmap='gray')
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[2].imshow(train_images[2], cmap='gray')
ax[2].set_xticks([])
ax[2].set_yticks([])
plt.show()
```
